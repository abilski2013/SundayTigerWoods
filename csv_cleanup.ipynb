```python
import pandas as pd
```

```python
df = pd.read_csv("static/data/water_new.csv")
```

```python
df = df.dropna(subset=['mean_value', 'max_value'])
df.head()
```

```python
county_groupby = df.groupby(["countyServed","analyteCode","year","mean_unit","waterSource"])
```

```python
county_df = county_groupby.mean().reset_index()
```

```python
county_df = county_df.to_csv("county_mean.csv")
```
